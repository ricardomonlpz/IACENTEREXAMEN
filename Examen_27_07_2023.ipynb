```python
# Muestra las primeras 5 filas del DataFrame
ds=df.copy()
print(ds.head(5))
```

```python
# @title Filtro y seleciona datos
# @markdown Seleciona los datos de la columna periodo mayor a 2018

filtro_periodo = ds['Periodo']> 2018
print("\nFilas donde la Periodo es mayor a 2018:")
print(ds[filtro_periodo])
```

```python
# Favor usar el nombre trimetre
# Pista la columna se llama "Unnamed: 1"
ds.rename(columns={"Unnamed: 1":"Trimestre"}, inplace=True)
print(ds.columns)
```

```python
# @title Calcular el total de ingresos para cada año y período

total_por_periodo = ds.groupby(['Periodo'])["Total"].sum()
print(total_por_periodo)
```

```python
valores_nulos = ds.isnull().sum()

porcentajes_nulos_ordenados= valores_nulos/ds.shape[0]*100

porcentajes_nulos_ordenados.sort_values(ascending=False)

porcentajes_nulos_ordenados
```

```python
import matplotlib.pyplot as plt

# Graficar los ingresos totales por período

plt.figure(figsize=(10, 6))
plt.bar(ds['Periodo'],ds['Total'])
plt.xlabel('Período')
plt.ylabel('Ingresos Totales')
plt.title('Ingresos Totales por Período')
plt.xticks()
plt.show()
```

# Convierte los salarios a dolares
### Supongamos que el tipo de cambio es 1 dólar = 20 unidades de la moneda local


```python
tipo_cambio=20
def convertir_a_dolares(salario):

    return salario / tipo_cambio

#df[].apply()

ds['Total_dolares'] = convertir_a_dolares(ds['Total'])
ds['1_salario_minimo_dolares'] = convertir_a_dolares(ds['1_salario_minimo'])
ds['2_salario_minimo_dolares'] = convertir_a_dolares(ds['2_salario_minimo'])
ds['3_salario_minimo_dolares'] = convertir_a_dolares(ds['3_salario_minimo'])
ds['5_salario_minimo_dolares'] = convertir_a_dolares(ds['5_salario_minimo'])

ds
```

```python
# @title Copia el token a en la repuesta de TEAMS
import hashlib

df_string = ds.to_string()

# Calcular el hash MD5
md5_hash = hashlib.md5(df_string.encode()).hexdigest()

print("Token del DataFrame:")
print(md5_hash)
```
